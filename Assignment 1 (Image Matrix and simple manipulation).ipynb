```python
#Using scipy to read the matrix of selected image

from scipy import ndimage
from scipy import misc

test_myimage = misc.imread('C:/Users/wicak/Desktop/test_image.jpeg')
type(test_myimage)
print(test_myimage)
```

```python
#using matplotlib to visualize the image

import matplotlib.pyplot as plt

plt.imshow(test_myimage)
plt.show()
```

```python
#Define 3 Dimensions in image using 'shape'
test_myimage.shape
```

```python
#slice array and access the RED component
redcomp=test_myimage[:,:,0]
```

```python
#slice array and access the GREEN component
greencomp=test_myimage[:,:,1]
```

```python
#slice array and access the BLUE component
bluecomp=test_myimage[:,:,2]
```

```python
print(redcomp)
plt.imshow(redcomp)
plt.show()
```

```python
print(greencomp)
plt.imshow(greencomp)
plt.show()
```

```python
print(bluecomp)
plt.imshow(bluecomp)
plt.show()
```

```python
#retrieve grayscale array for the image
#equation source : https://en.wikipedia.org/wiki/Grayscale#Converting_color_to_grayscale

grayimage = (0.2126 * redcomp) + (0.7152 * greencomp) + (0.0722 * bluecomp)

print(grayimage)
plt.imshow(grayimage)
plt.show()
```

```python
#.. and display 
plt.imshow(grayimage, cmap = plt.cm.Greys_r)
plt.show()
```

```python
#contrast manipulation
plt.imshow(grayimage, cmap=plt.cm.gray, vmin=50, vmax=60) #normal default value vmin = 0 and vmax=255
plt.show()
```
